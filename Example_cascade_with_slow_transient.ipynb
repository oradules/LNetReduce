```python
import lnetreduce

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# load the model
# REMARK : the exact layout in Figure 3 of the CMSB conference manuscript was designed manually 
# this is why it is not precisely reproduced in this notebook


filename = 'models/cascade_slow.csv'
G = lnetreduce.load_graph(filename)
lnetreduce.plot_graph(G)
```

```python
uG = lnetreduce.reduce_graph(G)
lnetreduce.plot_graph(uG)
```

```python
timescale = 13
np.random.seed(1)
plt.figure(figsize=(10,6))
plt.rcParams.update({'font.size': 22})
lnetreduce.simulate_and_plot(uG, timescale, title='Slow transient relaxation', method='odeint',ylabel='Amounts',initial_state=np.random.rand(6)*2)
plt.savefig('SlowRelaxation.png',dpi=300)
plt.show()
```
