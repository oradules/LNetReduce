```python
import lnetreduce
from matplotlib import pyplot as plt
```

```python
filename = "models/flower_7.csv"
```

```python
G = lnetreduce.load_graph(filename)
lnetreduce.plot_graph(G)
```

```python
%%time
for s in (35, 40, 50, 70):
    X = lnetreduce.simulate_and_plot(G, s, method='LSODA', title='LSODA - %s' % s)
    plt.show()
    X = lnetreduce.simulate_and_plot(G, s, method='Radau', title='Radau - %s' % s)
    plt.show()
    X = lnetreduce.simulate_and_plot(G, s, method='BDF', title='BDF - %s' % s)
    plt.show()
```

```python
uG = lnetreduce.reduce_graph(G)
lnetreduce.plot_graph(uG)
```

```python
%%time
for s in (35, 40, 50, 70):
    X = lnetreduce.simulate_and_plot(uG, s, method='LSODA', title='LSODA - %s' % s)
    plt.show()
    X = lnetreduce.simulate_and_plot(uG, s, method='Radau', title='Radau - %s' % s)
    plt.show()
    X = lnetreduce.simulate_and_plot(uG, s, method='BDF', title='BDF - %s' % s)
    plt.show()
```
