```python
import lnetreduce
```

```python
filename = "models/flower_CMSB_ERR.csv"
timescale = 10
```

```python
G = lnetreduce.load_graph(filename)
pos = lnetreduce.plot_graph(G)
```

```python
lnetreduce.simulate_and_plot(G, timescale)
```

```python
try:
    uG = lnetreduce.reduce_graph(G)
except:
    print("Error during the reduction")
```

```python
uG = lnetreduce.reduce_graph(G, partial=True)
_ = lnetreduce.plot_graph(uG, layout=pos)
```

```python
lnetreduce.simulate_and_plot(uG, timescale)
```
